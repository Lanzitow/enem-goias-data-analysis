```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Ler corretamente o CSV
df_resultados = pd.read_csv(
    "../dados/bruto/RESULTADOS_2024.csv",
    sep=";",
    encoding="latin1",
    low_memory=False
)

# 2️⃣ Filtrar Goiás
df_go = df_resultados[df_resultados["SG_UF_PROVA"] == "GO"].copy()

# 3️⃣ Criar Brasil (base completa)
df_br = df_resultados.copy()

# 4️⃣ Renomear colunas
df_go = df_go.rename(columns={
    "NU_NOTA_LC": "Linguagens",
    "NU_NOTA_MT": "Matemática",
    "NU_NOTA_CH": "Ciências Humanas",
    "NU_NOTA_CN": "Ciências da Natureza",
    "NU_NOTA_REDACAO": "Redação"
})

df_br = df_br.rename(columns={
    "NU_NOTA_LC": "Linguagens",
    "NU_NOTA_MT": "Matemática",
    "NU_NOTA_CH": "Ciências Humanas",
    "NU_NOTA_CN": "Ciências da Natureza",
    "NU_NOTA_REDACAO": "Redação"
})

# 5️⃣ Definir áreas
areas = [
    "Linguagens",
    "Matemática",
    "Ciências Humanas",
    "Ciências da Natureza",
    "Redação"
]

# 6️⃣ Remover NaN
df_go = df_go[areas].dropna()
df_br = df_br[areas].dropna()
```

```python
areas = [
    "Linguagens",
    "Matemática",
    "Ciências Humanas",
    "Ciências da Natureza",
    "Redação"
]

curtose_go = df_go[areas].kurt()
curtose_br = df_br[areas].kurt()

curtose_df = pd.DataFrame({
    "Curtose GO": curtose_go,
    "Curtose Brasil": curtose_br
})

print(curtose_df.round(2))
```

A curtose indica o grau de concentração das notas em torno da média. De modo geral, observa-se que as distribuições apresentam valores que sugerem presença de variações no formato dos dados, com algumas áreas apresentando maior concentração e outras maior dispersão. Isso evidencia que o comportamento das notas não é uniforme entre as áreas avaliadas

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Goiás
plt.figure(figsize=(6,6))

redacao_go = df_go["Redação"]
redacao_go = redacao_go[redacao_go > 0]

stats.probplot(redacao_go, dist="norm", plot=plt)

plt.title("QQ Plot — Redação (Goiás)")
plt.show()
```

O QQ plot da Redação indica que a distribuição das notas não segue perfeitamente uma distribuição normal. Observam-se desvios nas extremidades da reta de referência, sugerindo presença de assimetria e valores extremos. Esse comportamento é esperado, considerando as características específicas da avaliação da redação no ENEM.

```python
# Brasil
plt.figure(figsize=(6,6))

redacao_br = df_br["Redação"]
redacao_br = redacao_br[redacao_br > 0]

stats.probplot(redacao_br, dist="norm", plot=plt)

plt.title("QQ Plot — Redação (Brasil)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ordenar dados
dados_go = np.sort(df_go["Matemática"].dropna())
dados_br = np.sort(df_br["Matemática"].dropna())

# Percentis
p_go = np.linspace(0, 1, len(dados_go))
p_br = np.linspace(0, 1, len(dados_br))

plt.figure(figsize=(10,6))

# Goiás
plt.plot(p_go, dados_go, color="#1f77b4", linewidth=2, label="Goiás")
plt.scatter(p_go, dados_go, s=5, alpha=0.3)

# Brasil
plt.plot(p_br, dados_br, color="#ff7f0e", linewidth=2, label="Brasil")
plt.scatter(p_br, dados_br, s=5, alpha=0.3)

# Linha da mediana
plt.axvline(0.5, linestyle="--")

plt.title("Gráfico de Quantis — Matemática (Goiás vs Brasil)")
plt.xlabel("Quantil (p)")
plt.ylabel("Nota")
plt.legend()

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

O gráfico de quantis evidencia que as distribuições das notas de Matemática para Goiás e Brasil apresentam comportamento bastante semelhante ao longo de todos os percentis. As curvas praticamente sobrepostas indicam consistência no padrão de desempenho entre o estado e o cenário nacional. Observa-se que, na região central da distribuição, os valores são praticamente equivalentes, enquanto nas extremidades há presença de valores baixos e elevados, indicando assimetria na distribuição. O formato da curva sugere variação na concentração dos dados, reforçando que a distribuição não é perfeitamente uniforme.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Dados da Redação sem zeros
redacao_go = df_go["Redação"]
redacao_go = redacao_go[redacao_go > 0].dropna()

redacao_br = df_br["Redação"]
redacao_br = redacao_br[redacao_br > 0].dropna()

# Criar figura com dois gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# QQ Plot Goiás
stats.probplot(redacao_go, dist="norm", plot=axes[0])
axes[0].get_lines()[0].set_markerfacecolor("#1f77b4")
axes[0].get_lines()[0].set_markeredgecolor("#1f77b4")
axes[0].get_lines()[1].set_color("#1f77b4")
axes[0].set_title("QQ Plot — Redação (Goiás)")

# QQ Plot Brasil
stats.probplot(redacao_br, dist="norm", plot=axes[1])
axes[1].get_lines()[0].set_markerfacecolor("#ff7f0e")
axes[1].get_lines()[0].set_markeredgecolor("#ff7f0e")
axes[1].get_lines()[1].set_color("#ff7f0e")
axes[1].set_title("QQ Plot — Redação (Brasil)")

plt.tight_layout()
plt.show()
```
